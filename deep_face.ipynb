```python
from deepface import DeepFace
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
neutral=cv2.imread("/content/neutral.jpg")
plt.imshow(neutral)
plt.show()
```

```python
obj = DeepFace.analyze(neutral, actions = ['age','gender','race','emotion'])
print("your age is:",obj["age"],
      "your gender is:",obj["gender"],
      "your race is:",obj["race"],
      "your emotions are",obj["emotion"])
```
